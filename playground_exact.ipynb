```python
from plot_utils import *
from mat_utils import *
import numpy as np

# A = np.array([[2, 1], [1, 2]])
B = np.array([
    [1, 0, 1, 0],
    [0, 1, 0, 1]
]) * 2
C = B.T * 3
D = np.array([
    [5, 1, 0, 0],
    [1, 5, 1, 0],
    [0, 1, 5, 1],
    [0, 0, 1, 5]
])


J = bmat([[None, B], [C, D]]).astype(float).toarray()
plot_mat(J, log=False)
```

$B \cdot B^+ = I$

$C^+ \cdot C = I$

```python
B_inv = pinv_right(B)
C_inv = pinv_left(C)

B @ B_inv
```

```python
C_inv @ C
```

```python
C @ C_inv
```

```python
A_inv = -C_inv @ D @ B_inv
J_inv = bmat([
    [A_inv, C_inv],
    [B_inv, None]
]).toarray()
plot_mat(J_inv)
```
