```python
import numpy as np
from matplotlib import pyplot as plt
from mat_utils import *
from plot_utils import *
from mat_utils import inv as matrix_inv

def inv(x):
    try:
        x.shape[0]
    except:
        return 1 / x
    else:
        return matrix_inv(x)


J = np.array([
    [1, 2, 0, 3, 0, 4],
    [5, 6, 0, 0, 0, 7], 
    [0, 0, 8, 9, 0, 1], 
    [2, 0, 3, 4, 0, 0], 
    [0, 0, 0, 0, 0, 5], 
    [6, 7, 8, 0, 9, 1], 
]) * 10.

plot_mat(J, log=False)
```

# Making stabilization

```python
S44 = J[4, 4] - J[4, 5] * inv(J[5, 5]) * J[5, 4]
print(f'{S44 = }')

Jstab = J.copy()
Jstab[4, 4] = -S44
plot_mat(Jstab, log=False)
```

```python
plot_mat(inv(J))
```

```python
plot_mat(inv(Jstab), log=False)
```
