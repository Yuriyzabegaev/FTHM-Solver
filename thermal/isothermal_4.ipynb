```python
from itertools import product
import numpy as np
from matplotlib import pyplot as plt
from plot_utils import *
import porepy as pp
```

```python
g = 5
data = {
    'SAMG isothermal': load_data(f"../stats/stats_thermal_geo4h_steadyx{g}_sol{'SAMG'}_isothermal.json"),
    'SAMG thermal': load_data(f"../stats/stats_thermal_geo4h_steadyx{g}_sol{'SAMG'}.json"),
    'CPR isothermal': load_data(f"../stats/stats_thermal_geo4h_steadyx{g}_sol{'CPR'}_isothermal.json"),
    'CPR thermal': load_data(f"../stats/stats_thermal_geo4h_steadyx{g}_sol{'CPR'}.json"),
}

DAY = 24 * 60 * 60


for name, d in data.items():
    ts = np.cumsum(get_time_steps(d))
    ls = get_ls_indices(d)
    st, sl, op = get_num_sticking_sliding_open(d)
    total_frac = st + sl + op
    plt.plot(ts[ls] / DAY, sl / total_frac, label=name)

plt.ylabel('Share of sliding cells')
plt.xlabel('Simulation time, days')
plt.legend()
```
